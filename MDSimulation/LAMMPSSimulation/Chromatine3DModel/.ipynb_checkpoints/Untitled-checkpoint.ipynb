```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
  
image = mpimg.imread("images_trekyz.png")
X = np.asarray(image)
X = (X * 255).astype(np.uint8)
plt.imshow(X)
original_shape = X.shape 
```

```python
X_redi = np.reshape(X,(X.shape[0]*X.shape[1],X.shape[2]))
# Flatten the image to a 2D array where each row is a pixel and each column is a channel
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3).fit(X_redi)
```

```python
kmeans.labels_
plt.imshow(kmeans.cluster_centers_)
```

```python
reshaped_array = kmeans.labels_.reshape(original_shape[:2])
```

```python
plt.imshow(reshaped_array)
```
